```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,6))
```

#### Sleep Duration across age

```python
age_avg = df.groupby('age')['sleep_duration'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(x='age', y='sleep_duration', data=age_avg)
plt.title("Average Sleep Duration Across Age", fontsize=14, fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Average Sleep Duration (hours)")
plt.show()
```

#### Quality of VS BMI_ Category

```python
plt.figure(figsize=(10,6))
sns.barplot(
    y="bmi_category",  # switch x/y for horizontal
    x="quality_of_sleep",
    data=df,
    palette="Set2"
)
plt.title("Quality of Sleep by BMI Category", fontsize=14, fontweight="bold")
plt.xlabel("Quality of Sleep")
plt.ylabel("BMI Category")
plt.show()
```

```python
sleep_counts = df['sleep_disorder'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    sleep_counts,
    labels=sleep_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Set2")
)
plt.title("Distribution of Sleep Disorders", fontsize=14, fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='sleep_duration', data=df, alpha=0.6, color="skyblue")
sns.regplot(x='age', y='sleep_duration', data=df, scatter=False, color='red', line_kws={'linewidth':2, 'linestyle':'--'})
plt.title("Sleep Duration Across Age (Trend)", fontsize=14, fontweight="bold")
plt.xlabel("Age")
plt.ylabel("Sleep Duration (hours)")
plt.show()
```
